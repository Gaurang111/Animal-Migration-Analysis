```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

# Temperature Data Section
- Creating a line graph of the temperature

```python
#create a local path for the CSVs
os.chdir(r'C:\Users\kenne\OneDrive\Documents\canadian_temp')

#reading in a CSV about the weather conditions in multiple locations in Canada
df_1 = pd.read_csv('Canadian_climate_history.csv')

#making the data set smaller to be more mangeable as the first 825 rows are not necessary
df_1 = df_1[825:]

#dropping unnecessary columns to make the data set more mangeable
df_1 = df_1.drop(df_1.columns[2:9], axis = 1)

df_1.head()
```

```python
#importing the datetime module to allow changing the format of the date
import datetime as dt

#changing the Local_Date column to a datetime type of variable
df_1['LOCAL_DATE'] = pd.to_datetime(df_1['LOCAL_DATE'])

#formatting the Local_Date column to MM-YYYY format
df_1['LOCAL_DATE'] = df_1['LOCAL_DATE'].dt.strftime('%m-%Y')

df_1.head()
```

```python
#filtering out the Local_Date rows so that only the row for January remains
#the average temperature for January will be calculated
df_1 = df_1[df_1['LOCAL_DATE'].str.startswith('01')]

df_1.head()
```

```python
#removing the columns so that only the necessary columns, Local_Date and Mean_Temperature_Quebec, remain
df_1 = df_1[['LOCAL_DATE', 'MEAN_TEMPERATURE_QUEBEC']]
```

```python
#dropping any null values
df_1 = df_1.dropna()

#changing the Local_Date column back to a string type, and remove the month from the date
df_1['LOCAL_DATE'] = df_1['LOCAL_DATE'].astype(str).str.replace('01-', '')
```

```python
#groupby each year in the Local_Date column by the average of the Mean_Temperature_Quebec for each day in January
dc_1 = df_1.groupby(['LOCAL_DATE'])['MEAN_TEMPERATURE_QUEBEC'].mean()

#convert to a Data Frame and remove rows so that only the years 1970 and onward are left
#only 1970 and onward for simplicity
dc_1 = pd.DataFrame(dc_1)[26:]

dc_1.head()
```

```python
#showing the plot in the notebook
%matplotlib inline

#showing the temperature values per year in a line graph with the color blue and little circle markers
plt.plot(dc_1,'b', marker = 'o')

#the x-axis is partitioned into groups of 10
plt.xticks([0, 10, 20, 30, 40, 50])

#creating a title for the graph
plt.title('Average Temperature in Canada (Quebec) From January 1970-2019')

#the x-axis is labelled for the years
plt.xlabel('Year')

#the y-axis is labelled for the range of the temperature
plt.ylabel('Temperature (C)')

plt.show()
```

# Lemming Population Data Section
- Creating a bar graph of lemming populations

```python
#create a local path for the CSVs
os.chdir(r'C:\Users\kenne\OneDrive\Documents')

#reading in a fwf format file
#skip the first row (which is the header)
#colspecs: only keeping the year column as each entry is a lemming nest
df_1 = pd.read_fwf('new_lemming_pop.txt', skiprows = 1, colspecs = [(0, 4)], header = None)

#renaming the only column to 'Year'
df_2 = df_1.rename(columns = {0:'Year'})

df_2.head()
```

```python
#calculating the number of rows per year and then sorting in order
#converting to a Data Frame
df_3 = pd.DataFrame(df_2.value_counts()).sort_index()

df_3.head()
```

```python
#showing the plot in the notebook
%matplotlib inline

#arranging the year values by hand for simplicity
x_value = [2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]

#arranging the lemming nest values by hand for simplicity
y_value = [263, 229, 103, 570, 673, 144, 52, 395, 582, 399, 250, 128, 489]

#creating a bar graph of the year and number of lemming nests
#changing the color to cyan and adding a blue edge
plt.bar(x_value, y_value, color = 'cyan', edgecolor = 'blue')

#x-axis is labelled 'Year'
plt.xlabel('Year')

#y-axis is labelled 'Number of Lemming Nests'
plt.ylabel('Number of Lemming Nests')

#adding a title
plt.title('Number of Lemming Nests Recorded per Year in Bylot Island')

plt.show()
```
