```python
data = pd.read_csv('/content/drive/MyDrive/DataSciencePro/Row Data/sea_ice_chart.csv', header = None, names=['Year',
                                                                                             'Foxe Basin (thousands of square kilometres)',
                                                                                             'Kane Basin (thousands of square kilometres)',
                                                                                             'Baffin Bay (thousands of square kilometres)',
                                                                                             'Beaufort (thousands of square kilometres)',
                                                                                             'Canadian Arctic Archipelago (thousands of square kilometres)',
                                                                                             'Hudson Bay (thousands of square kilometres)',
                                                                                             'Hudson Strait (thousands of square kilometres)',
                                                                                             'Davis Strait (thousands of square kilometres)',
                                                                                             'North Labrador Sea (thousands of square kilometres)'])
```

```python
data.head(100)
```

```python
# selects the third column
df = data.iloc[3:]
#df = df.astype(int)
```

```python
df.head()
```

```python
# Removing the NAN values, unnecessary string types and renumbering to new index
df = df.apply(pd.to_numeric , errors='coerce')
df = df.dropna()
df = df.reset_index(drop=True)
```

```python
# Select value from dataset with new index
df.head(10)
```

```python
# New dataframe after cleaning the data
df = pd.DataFrame(df)
```

```python
# selecting only two column from dataframe "Year" and "Baffin Bay" and converting into Int Data type
df = df[['Year', 'Baffin Bay (thousands of square kilometres)']].astype(int).apply(lambda x: x.replace('.0',''))
print(df)
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# plotting the bar in normal view 

df = pd.DataFrame(df, columns=['Year', 'Baffin Bay (thousands of square kilometres)'])
fig = plt.figure(figsize =(20, 10))

plt.bar(df["Year"], df["Baffin Bay (thousands of square kilometres)"], color = 'maroon', width = 0.6 )
plt.title('Sub-region sea ice area trends, Northern Canadian Waters, 1968 to 2020', fontsize=16, color = 'blue')
plt.xlabel('Year', fontsize=16, color = 'blue')
plt.ylabel('Baffin Bay Square Kilometres', fontsize=16, color = 'blue')
plt.grid(True)
plt.show()
```
